```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'mtcars.csv'
data = pd.read_csv(url)
data.head(21)
mpg= data['mpg']
disp = data['disp']
hp = data['hp']
wt = data['wt']

data.hist()
plt.show()
correlation= data.corr()  #Độ tương quan giữa các indexindex
print(correlation)
```

```python
x=data.drop(data.columns[[0,2]],axis=1)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model as lm

# Tạo biến dựa trên những atribute cảu data
X = pd.DataFrame(data['wt'])
y = data['mpg']

#Tạo một model theo dạng hàm bậc 1
model = lm.LinearRegression()
#Cho phép giá trị cảu data được đưa về dạng bậc 1 tuyến tính
model.fit(X,y)
print(model.intercept_)
print(model.coef_)
# độ chính xác
model.score(X,y)
```

```python
y_pred = model.predict(X)
residual = data['mpg']-y_pred
residual.describe()
```

```python
#Vẽ đồ thị dưới dạng biểu đồ chấm
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9,9),color = "black")
#Vẽ chồng đồ thị dưới dạng đường thẳng để so sánh giữa biểu đồ thực và giá trị ước lượng y_pred
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model_2 = lm.LinearRegression()
x1 = pd.DataFrame([data['wt'],data['wt']**2]).T
y1 = data['mpg']
model_2.fit(x1,y1)

print('Intercept cua mo hinh bac hai:')
print(model_2.intercept_)
print('coefficient cuar mo hinh bac hai')
print(model_2.coef_)
print('Do chinh xac cua mo hinh')
print(model_2.score(x1,y1))
```

```python
range = np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range,range**2]).T
y_pred = model_2.predict(X=predictor)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9),color = 'black', xlim=(0,7))
plt.plot(range,y_pred,color = 'red')
```
