```python
import numpy as np
import skfuzzy as fuzz 
from skfuzzy import control as ctrl 
```

```python
tocdogio = ctrl.Antecedent(np.arange(0.7,3.2,0.1),'tocdogio')
nhietdo = ctrl.Antecedent(np.arange(20.4,41.3,1),'nhietdo')
luongmua = ctrl.Consequent(np.arange(0.2,9.2,0.1),'luongmua')
```

```python
tocdogio['Very Low WS']= fuzz.trimf(tocdogio.universe,[0.7,0.7,1.2])
tocdogio['Low WS']= fuzz.trimf(tocdogio.universe,[0.7,1.2,1.5])
tocdogio['Normal WS']= fuzz.trimf(tocdogio.universe,[1.2,1.5,2.5])
tocdogio['High WS']= fuzz.trimf(tocdogio.universe,[1.5,2.5,3.1])
tocdogio['Very High WS']= fuzz.trimf(tocdogio.universe,[2.5,3.1,3.1])

nhietdo ['Very Low TP']= fuzz.trimf(nhietdo .universe,[20.4,20.4,25])
nhietdo ['Low TP']= fuzz.trimf(nhietdo .universe,[20.4,25,30])
nhietdo ['Normal TP']= fuzz.trimf(nhietdo .universe,[25,30,35])
nhietdo ['High TP']= fuzz.trimf(nhietdo .universe,[30,35,41.2])
nhietdo ['Very High TP']= fuzz.trimf(nhietdo .universe,[35,41.2,41.2])

luongmua['Very Low RF']= fuzz.trimf(luongmua.universe,[0.9,0.9,1.8])
luongmua['Low RF']= fuzz.trimf(luongmua.universe,[0.9,1.8,4])
luongmua['Normal RF']= fuzz.trimf(luongmua.universe,[1.8,4,7])
luongmua['High RF']= fuzz.trimf(luongmua.universe,[4,7,9.1])
luongmua['Very High RF']= fuzz.trimf(luongmua.universe,[7,9.1,9.11])


tocdogio.view()
nhietdo.view()
luongmua.view()
```

```python
rule1 = ctrl.Rule(tocdogio['Very Low WS']& nhietdo['Very Low TP'], luongmua['Very Low RF'])
rule2 = ctrl.Rule(tocdogio['Low WS']& nhietdo['Low TP'], luongmua['Very Low RF'])
rule3 = ctrl.Rule(tocdogio['Normal WS']& nhietdo['Normal TP'], luongmua['Normal RF'])
rule4 = ctrl.Rule(tocdogio['Normal WS']& nhietdo['Very Low TP'], luongmua['Normal RF'])
rule5 = ctrl.Rule(tocdogio['Normal WS']& nhietdo['High TP'], luongmua['Low RF'])
rule6 = ctrl.Rule(tocdogio['Normal WS']& nhietdo['Low TP'], luongmua['High RF'])
rule7 = ctrl.Rule(tocdogio['Normal WS']& nhietdo['Very High TP'], luongmua['Low RF'])
rule8 = ctrl.Rule(tocdogio['Very Low WS']& nhietdo['Normal TP'], luongmua['Low RF'])
rule9 = ctrl.Rule(tocdogio['Low WS']& nhietdo['Normal TP'], luongmua['Low RF'])
rule10 = ctrl.Rule(tocdogio['High WS']& nhietdo['Normal TP'], luongmua['Low RF'])
rule11= ctrl.Rule(tocdogio['Very High WS']& nhietdo['Normal TP'], luongmua['Normal RF'])
rule12 = ctrl.Rule(tocdogio['Low WS']& nhietdo['Very Low TP'], luongmua['High RF'])
rule13 = ctrl.Rule(tocdogio['High WS']& nhietdo['Very Low TP'], luongmua['Very Low RF'])
rule14 = ctrl.Rule(tocdogio['Very High WS']& nhietdo['Very Low TP'], luongmua['Very High RF'])
rule15 = ctrl.Rule(tocdogio['Very Low WS']& nhietdo['Low TP'], luongmua['Normal RF'])
rule16= ctrl.Rule(tocdogio['High WS']& nhietdo['Low TP'], luongmua['Normal RF'])
rule17 = ctrl.Rule(tocdogio['Very High WS']& nhietdo['Low TP'], luongmua['Very High RF'])
rule18 = ctrl.Rule(tocdogio['Very Low WS']& nhietdo['High TP'], luongmua['Low RF'])
rule19 = ctrl.Rule(tocdogio['Low WS']& nhietdo['High TP'], luongmua['Very Low RF'])
rule20 = ctrl.Rule(tocdogio['High WS']& nhietdo['High TP'], luongmua['Low RF'])
rule21= ctrl.Rule(tocdogio['Very High WS']&nhietdo['High TP'], luongmua['Low RF'])
rule22 = ctrl.Rule(tocdogio['Very Low WS']&nhietdo['Very High TP'], luongmua['Very Low RF'])
rule23 = ctrl.Rule(tocdogio['Low WS']&nhietdo['Very High TP'], luongmua['Low RF'])
rule24 = ctrl.Rule(tocdogio['High WS']&nhietdo['Very High TP'], luongmua['Very Low RF'])
rule25 = ctrl.Rule(tocdogio['Very High WS']&nhietdo['Very High TP'],luongmua['Very High RF'])



luongmua_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule9,rule20,rule21,rule22,rule23,rule24,rule25])
Rainfalling= ctrl.ControlSystemSimulation(luongmua_ctrl)
Rainfalling.input['nhietdo']= 28
Rainfalling.input['tocdogio']= 9
Rainfalling.compute()
print(Rainfalling.output['luongmua'])
luongmua.view(sim=Rainfalling)
```
